# 01 - 数据探索

探索 MLS 点云数据，了解数据结构和质量。

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt

# 查看数据目录
from pathlib import Path
data_root = Path('../data')
print('数据目录结构:')
for p in sorted(data_root.rglob('*')):
    if p.is_file():
        print(f'  {p.relative_to(data_root)}')
```

```python
# 加载一个点云文件查看
import laspy

sample_files = list(Path('../data/00_raw').glob('*.las'))
if sample_files:
    las = laspy.read(str(sample_files[0]))
    points = np.vstack([las.x, las.y, las.z]).T
    print(f'点数: {len(points):,}')
    print(f'坐标范围:')
    print(f'  X: [{points[:,0].min():.2f}, {points[:,0].max():.2f}]')
    print(f'  Y: [{points[:,1].min():.2f}, {points[:,1].max():.2f}]')
    print(f'  Z: [{points[:,2].min():.2f}, {points[:,2].max():.2f}]')
```

```python
# 可视化点云 (2D投影)
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.scatter(points[::10, 0], points[::10, 1], s=0.1, alpha=0.5)
plt.title('XY 平面 (俯视图)')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.axis('equal')

plt.subplot(1, 3, 2)
plt.scatter(points[::10, 0], points[::10, 2], s=0.1, alpha=0.5)
plt.title('XZ 平面 (侧视图)')
plt.xlabel('X (m)')
plt.ylabel('Z (m)')

plt.subplot(1, 3, 3)
plt.hist(points[:, 2], bins=50, alpha=0.7)
plt.title('Z坐标分布 (高度)')
plt.xlabel('Z (m)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
