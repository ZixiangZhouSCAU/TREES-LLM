TreeVoxel Tokenizer 和 TREES-LLM 模型训练与测试

```python
import sys
sys.path.insert(0, '..')

import torch
import numpy as np

from src.models.tree_voxel_tokenizer import TreeVoxelTokenizer
from src.models.tree_part_attention import TreePartAttention
from src.models.projection_layer import PointCloudProjectionLayer
```

```python
# 测试 TreeVoxel Tokenizer
tokenizer = TreeVoxelTokenizer(
    num_geo_tokens=2048,
    num_sem_tokens=2048,
    token_dim=128,
)

# 模拟输入: batch=2, 4096 points, 3D坐标
points = torch.randn(2, 4096, 3)

with torch.no_grad():
    outputs = tokenizer(points)

print('TreeVoxel Tokenizer 输出:')
print(f'  几何token索引: {outputs["geo_tokens"].shape}')
print(f'  语义token索引: {outputs["sem_tokens"].shape}')
print(f'  总loss: {outputs["total_loss"].item():.4f}')
```

```python
# 测试 TreePartAttention
attention = TreePartAttention(dim=512, num_heads=8)

features = torch.randn(2, 4096, 512)
coords = torch.randn(2, 4096, 3)

out = attention(features, coords)
print(f'TreePartAttention 输出: {out.shape}')
```
